1.What is a parameter?

Ans.A parameter is a numerical value that describes a characteristic of an entire population in statistics.

Examples of parameters include:

Population mean (μ)
Population standard deviation (σ)
Population proportion (p)

Parameters are usually fixed but unknown, and they are estimated using sample data (through statistics like sample mean and sample standard deviation).

In simple terms, a parameter represents the true value of a population, while a statistic is an estimate of that parameter based on a sample.

4.How does loss value help in determining whether the model is good or not?

Ans.The loss value measures how far the predictions of a machine learning model are from the actual (true) values. It is a numerical indicator of the model’s error.

A high loss value means the model is making large errors and is not performing well.
A low loss value means the model’s predictions are closer to the actual values and it is performing better.

During training, the model tries to minimize the loss function to improve its accuracy. If the loss keeps decreasing over time, it indicates that the model is learning effectively.

However, if the loss becomes too low on training data but high on test data, it may indicate overfitting.

6.How do we handle categorical variables in Machine Learning? What are the common techniques?

Ans.Categorical variables represent data in the form of labels or groups (e.g., gender, city, color). Since machine learning models work with numerical data, these variables must be converted into numeric form before training.

Common Techniques to Handle Categorical Variables:
Label Encoding
Assigns a unique number to each category.
Example: Male = 0, Female = 1
Best for ordinal data (where order matters).
One-Hot Encoding
Creates separate binary columns for each category (0 or 1).
Example: Red, Blue, Green → three columns.
Best for nominal data (no natural order).
Ordinal Encoding
Used when categories have a meaningful order.
Example: Low = 1, Medium = 2, High = 3.
Binary Encoding (Advanced)
Converts categories into binary digits to reduce dimensionality.

7.What do you mean by training and testing a dataset?

Ans.In machine learning, a dataset is usually divided into two parts: training data and testing data.

Training Dataset:

The training dataset is used to teach the model. The model learns patterns, relationships, and structures from this data by adjusting its parameters.

Testing Dataset:

10.How do we split data for model fitting (training and testing) in Python? How do you approach a Machine Learning problem?

Ans.In Python, data is split into training and testing sets using Scikit-learn (Scikit-learn). The most common method is train_test_split().

Example:

# Example dataset
X = [[1], [2], [3], [4], [5], [6]]

y = [10, 20, 30, 40, 50, 60]

# Split data (80% train, 20% test)

test_size=0.2 → 20% data is used for testing
random_state → ensures same split every time
Training data is used to build the model, testing data is used for evaluation.

16.What is an Optimizer? What are different types of optimizers? Explain each with an example.

Ans.An optimizer is an algorithm used in machine learning and deep learning to reduce the loss function by adjusting the model’s parameters (weights and biases). Its main goal is to improve model accuracy by finding the minimum error.

Optimizers are a key part of training models in machine learning frameworks like Scikit-learn (Scikit-learn) and deep learning libraries.

Types of Optimizers:
1. Gradient Descent (Batch Gradient Descent)
Uses the entire dataset to update weights.
Slow but stable.

Example idea:
Updating weights after calculating error from all training samples.

2. Stochastic Gradient Descent (SGD)
Updates weights after each training sample.
Faster but less stable (more fluctuations).

Example idea:
Updating model after every single student’s marks in a dataset.

3. Mini-Batch Gradient Descent
Combines both batch and SGD.
Uses small groups (batches) of data for updates.
Most commonly used in practice.

Example idea:
Updating weights after every 32 or 64 samples instead of all or one.

4. Momentum Optimizer
Improves SGD by adding “memory” of past updates.
Helps move faster in correct direction and reduces oscillations.

5. Adam Optimizer (Adaptive Moment Estimation)
Most popular optimizer in deep learning.
Combines Momentum + adaptive learning rate.
Works well for most problems.

22.How do we perform scaling in Python?

Ans.In Python, feature scaling is performed using Scikit-learn (Scikit-learn), which provides tools like StandardScaler and MinMaxScaler.

Scaling is an important preprocessing step to normalize feature values before training a machine learning model.

1. Standardization (Using StandardScaler)

This method transforms data so that it has:

Mean = 0
Standard deviation = 1
from sklearn.preprocessing import StandardScaler

# Sample data
data = [[10], [20], [30], [40], [50]]

# Create scaler
scaler = StandardScaler()

# Fit and transform data
scaled_data = scaler.fit_transform(data)

print(scaled_data)
2. Normalization (Using MinMaxScaler)

This method scales data between 0 and 1.

from sklearn.preprocessing import MinMaxScaler

# Sample data
data = [[10], [20], [30], [40], [50]]

# Create scaler
scaler = MinMaxScaler()

# Fit and transform data
scaled_data = scaler.fit_transform(data)

print(scaled_data)

Explanation:
fit_transform() learns the data parameters and applies scaling.
StandardScaler is used when data follows a normal distribution.
MinMaxScaler is used when data needs to be bounded between 0 and 1.

25.Explain data encoding?

Ans.Data encoding is the process of converting categorical (text-based) data into numerical form so that machine learning models can understand and process it.

Since most ML algorithms work only with numbers, encoding is an essential preprocessing step.

Why data encoding is needed:
Machine learning models cannot directly process text data.
Converts categories into a numerical format.
Improves model performance and compatibility.
Common Types of Data Encoding:
Label Encoding
Assigns a unique number to each category.
Example:
Red → 0
Blue → 1
Green → 2
Best for ordinal data (where order matters).
One-Hot Encoding
Creates separate binary columns for each category.
Example: Red, Blue, Green → three columns (0/1 values).
Best for nominal data (no order).
Ordinal Encoding
Used when categories have a meaningful order.
Example: Low → 1, Medium → 2, High → 3
